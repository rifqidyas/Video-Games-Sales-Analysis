```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/vgsales.csv", index_col=0)
df.head(15)
```

```python
df.info()
df.shape
df.describe()
```

```python
df.describe(include='object')
```

```python
print("missing values sebelum cleaning")
df.isnull().sum()
```

```python
df['Year'] = df['Year'].fillna(df['Year'].median())
df['Publisher'] = df['Publisher'].fillna('Unknown')
df = df.dropna()
df["Year"] = df['Year'].astype(int)
```

```python
print("missing value setelah cleaning")
df.isnull().sum()
```

```python
df[df.duplicated(keep=False)]
```

```python
df = df.drop_duplicates()
```

```python
df.to_csv("../data/processed/vgsales_cleaned.csv", index=False)
```
